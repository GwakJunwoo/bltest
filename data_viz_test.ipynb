```python
import data_viz as viz
import plotly.express as px
import pandas as pd
```

```python
df_line = pd.read_csv('sample.csv', encoding='EUC-KR')

def cum_profit(df):
    pct = df.pct_change()
    cum_pct = (1+pct).cumprod()-1
    cum_pct.iloc[0] = [0 for k in range(len(cum_pct.columns))]
    return cum_pct

df_line=df_line.set_index('Date')
df_line =cum_profit(df_line)

df_line.head()
```

```python
viz.line_plot(df_line, x=df_line.index, y=['ACWI', 'BOND', 'USO'], title="<b>Time-series<b>", data_source='Yahoo Finance')
```
